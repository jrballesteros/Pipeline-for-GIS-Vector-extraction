```python
# Este script crea un dataset con solapamiento de (5%, 10%, 20%) como parámetro, parte
# la imagen en chips cuadrados de un tamaño especificado

import os
import cv2

path_to_img = r"C:\Users\john\Downloads\AutoMap\datos\El_Retiro.tif"
#path_to_img = r"C:\Users\john\Downloads\AutoMap\datos\1px\La_Ceja_FColor_Clip_RGDSM.tif"
#path_to_img = r"C:\Users\john\Downloads\AutoMap\datos\Infer_Ortom.tif"

savedir = "C:/Users/john/Downloads/AutoMap/datos/img256_3m_RGB/"
#savedir = "C:/Users/john/Downloads/AutoMap/datos/mos256_DSM_50cm_RGB_overlap20/"
#savedir = "C:/Users/john/Downloads/AutoMap/datos/datasetoverlap/"

path_to_msk = r"C:\Users\john\Downloads\AutoMap\datos\El_Retiro_roads_3m.tif"

savedir_msk = "C:/Users/john/Downloads/AutoMap/datos/msk256_3m_RGB/"

split_width = 256
split_height = 256
#split_width = 600
#split_height = 600

overlap = 0.3

def start_points(size, split_size, overlap):
    points = [0]
    stride = int(split_size * (1-overlap))
    counter = 1
    while True:
        pt = stride * counter
        if pt + split_size >= size:
            points.append(size - split_size)
            break
        else:
            points.append(pt)
        counter += 1
    return points

count = 1
name = 's'
frmt = 'png'

img = cv2.imread(path_to_img)
img_h, img_w, _ = img.shape

X_points = start_points(img_w, split_width, overlap)
Y_points = start_points(img_h, split_height, overlap)
for i in Y_points:
    for j in X_points:
        split = img[i:i+split_height, j:j+split_width]
        cv2.imwrite(os.path.join(savedir + '{}{}.{}'.format(name, count, frmt)), split)
        count += 1

msk = cv2.imread(path_to_msk)
msk_h, msk_w, _ = msk.shape

count = 1
X_points = start_points(msk_w, split_width, overlap)
Y_points = start_points(msk_h, split_height, overlap)
for i in Y_points:
    for j in X_points:
        split = msk[i:i+split_height, j:j+split_width]
        cv2.imwrite(os.path.join(savedir_msk + '{}{}.{}'.format(name, count, frmt)), split)
        count += 1

print("Se han producido ", count, "img-msk con ",  overlap*100, "% de overlap")
```
